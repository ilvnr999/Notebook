git中檔案的狀態
* Untracked 版本加入暫存區後(add)才加入的檔案 未被git追蹤管控

* Tracked 被追蹤的狀態 已加入暫存區(add)

    * Modified 修改過後尚未加入暫存區(add)

    * Staged 修改後已加入暫存區(add)仍未提交(commit)

    * Commited 修改後已加入暫存(add)也已提交(commit)

### 建立一個資料夾後

1. ```git init``` 資料夾會新增一個.git的版控資料夾

1. ```git add```  檔案加入暫存區 檔案staged狀態

2. ```git commit -m " "``` 檔案加入local repository

3. ```git push``` 檔案加入remote repository

### 從remote

* ```git clone url/ssh``` 克隆一個專案

* ```git pull``` 將local更新至remote的狀態

    = ```git fetch``` + ```git merge```


### 檢視

* ```git status``` 檢視所在分支 origin分支 檔案更改狀況

* ```git log``` 顯示commit歷史紀錄

    ```git log --oneline``` 顯示簡易版本

* ```git reflog``` 查看local所有git操作記錄

* ```git diff``` 得知工作區與暫存區的不同之處

### 改動

* ```git reset``` 依版本改動暫存區與工作區

    * 默認 將加入暫存區的丟回工作區 Modified狀態

    * ```--hard commit ID``` 根據commit ID回覆到指定版本

      重置HEAD 丟棄暫存區與工作區
    
    * ```--soft commit ID```

      只重置HEAD 保留暫存區與工作區

* ```git revert HEAD```

    * ```HEAD~``` 上一個提交紀錄 ```~2``` 上兩個 以此類推

### Branch

* ```git branch``` 查看分支

* ```git branch 分支名稱``` 建立分支

* ```git checkout 分支名稱``` 切換至分支

* ```git checkout -b 分支名稱``` 快速建立與切換分支

* ```git branch -d 分支名稱``` 安全刪除分支 刪除以合併的分支(需離開此分支才可以刪除)

* ```git branch -D 分支名稱``` 強制刪除分支

* 刪除後的branch 可以透過 ```git checkout commit ID``` 然後 ```git checkout -b 分支名稱``` 恢復原本的分支

    或是 ```git branch 分支名稱 commit ID```

* ```git push --set-upstream origin 分支名稱``` 將local的分支推上remote

### merge

1. ```git checkout 要合併到的分支```

2. ```git merge 要被合併的分支```

    * treminal會開啟一個編輯器 在內部輸入完commit message後按ESC離開 再輸入:wq離開編輯器

    * 合併後會在合併入的分支上面新增一個節點 被合入的分支仍維持原狀態 除非也反過來merge

### 回復merge

1. ```git checkout 合併入的分支```

2. ```git log --oneline``` 找到合併前合併入的分支合併前的commit ID

3. ```git reset --hard commit ID``` 恢復合併

    ```git reset --hard ORIG_HEAD``` 利用ORIG_HEAD回到上個版本

### 快轉合併

若main分支出去後 在分之上修改 main沒有做任何改動

* ```git merge 要合併的分支``` 會直接將main快轉至分之的狀態 不會有耳朵的分支合併

* ```git merge --no-ff 合併的分支``` 這就會長出耳朵的分支合併
