##比較loc與iloc

loc  (location)          使用標籤指向 row column cell儲存格

iloc (integer location)  使用索引

```python
fruit = {
    'orange':[5, 7, 0, 1, 20],
    'apples':[2, 3, 7, 2, 12],
    'grapes':[1, 2, 6, 0, 15],
    'pear':[0, 13, 25, 1, 32]
}
df = pd.DataFrame(fruit, index=['April', 'May', 'June', 'July', 'August'])
df
```

單一位置

```python
print(df.loc['April'])
print(df.iloc[0])
```

取得範圍

```python
print(df.loc['April':'June'])
print(df.iloc[0:3])
```

取得特定行列

```python
print(df.loc['April':'June','apples':'grapes'])
print(df.iloc[0:3,1:3])
```

使用=更新值

```python
print(df['orange']==5)  # 判斷每個row使否有符合判別式
```

```python
df.loc[df['orange']==5,['orange']] = 50
df
```
