#Pandas 基礎

```import pandas as pd```

### series

一維 本質為一個column

* ```df = pd.Series([list], index=[list])``` 建立Series

* ```df[index/'name']``` 透過index 或是 name 讀取資料

### DataFrame

二維 由Series集合組成的多維表格

透過list建立

* ```df = pd.DataFrame([[list1], [list2]], index=[], columns=[])```

  一個list為一個row建立 index 與 columns 可省略

透過dict與list建立

* ```df = pd.DataFrame({'column1' : ['Bob', 'Amy'], 'column2' : [23, 25]}, index=[])```

  dict一個key:value為一個column建立

  index 可省略

透過list與dict建立

* ```df = pd.DataFrame([{'Name': 'Bob', 'Age': 23}, {'Name': 'Amy', 'Age': '25'}], index=[])```

  一個dict為一個row建立 key為column name

  index 可省略

取值

* ```df.loc[index_name]``` 透過名稱讀取row Series type

* ```df.iloc[index]``` 透過索引讀取row Series type

* ```df[column_name]``` 透過名稱讀取column Series type

* ```df[[column_name]]``` 透過名稱讀取column DataFrame type

* ```df.loc[:, name]``` 透過名稱使用切片讀取cloumn Series type

* ```df.iloc[:, column_index]``` 透過column索引使用切片讀取 column Series type

* ```df.loc[df['column_name1'] == 'value', ['column_name2', 'column_name3']]``` 透過判別式取出沒有index的DataFrame某個列1為某值的 列2跟列3的值

### I/O

* ```df = pd.read_csv('file')``` 讀取csv檔

* ```df = pd.read_excel('file')``` 讀取excel檔

* ```df = pd.read_json('file')``` 讀取json檔

* ```df = df.to_csv('file_name',index)``` 寫入csv index是否保留True放在新的一行False拋棄

* ```df.to_excel('file_name',sheet_name=,index)``` 寫入excel

* ```df.to_json('file_name')``` 寫入json

### 操作

* ```df.shape``` 取得行列數量 回傳Tuple

* ```df.to_string()``` 文件完整顯示

* ```df.head()``` 顯示前5個row 括號中可填入數字設定

* ```df.tail()``` 顯示後5個row

* ```df.describe()``` 分析統計指標  ```(percentiles=[0.3, 0.5,0.7])``` 修改百分比

* ```df.T``` 轉置

* ```df.info()``` 資料理解 幾個column 各名稱 多少非0 資料型態

* ```df.columns``` 取得所有column name

* ```df.isnull()``` 檢查缺失值 回傳一個與df一樣大的DataFrame內容為bool

* ```df.isnull().sum()``` 列出每個column個別有多少個缺失值

* ```df.isnull().sum().sum()``` 計算全部總共多少個缺失值

* ```df[df.index.isin(range(2, 10))]``` 選擇特定範圍的row
